```python
import numpy as np
import pandas as pd
data=pd.read_csv('hypothyroid.csv')
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
for x in data.columns:
    data[x]=enc.fit_transform(data[x])
data.info()
```

```python
data.head()
```

```python
data=data.drop_duplicates()
data.info()
```

```python
data.describe()
```

```python
data['age']=(data['age']-data['age'].min())/(data['age'].max()-data['age'].min())
data['TT4']=(data['TT4']-data['TT4'].min())/(data['TT4'].max()-data['TT4'].min())
data['T4U']=(data['T4U']-data['T4U'].min())/(data['T4U'].max()-data['T4U'].min())
data['FTI']=(data['FTI']-data['FTI'].min())/(data['FTI'].max()-data['FTI'].min())
```

```python
data.head()
```

```python
y=data['binaryClass']
x=data.drop(['binaryClass'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.1,stratify=y)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from scipy import sparse
from scipy.sparse import csr_matrix
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(xtrain, ytrain)
logreg_predict = logreg.predict(xtest)
logreg_acc = accuracy_score(logreg_predict, ytest)
print("Test accuracy:{:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(ytest, logreg_predict))
print("\n")
print(classification_report(ytest, logreg_predict))
```

```python
dt =  DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
dt_predict = dt.predict(xtest)
dt_acc = accuracy_score(dt_predict, ytest)
print("Test accuracy:{:.2f}%".format(dt_acc*100))
```

```python
print(confusion_matrix(ytest, dt_predict))
print("\n")
print(classification_report(ytest, dt_predict))
```

```python
knn = KNeighborsClassifier()
knn.fit(xtrain, ytrain)
knn_predict = knn.predict(xtest)
knn_acc = accuracy_score(knn_predict, ytest)
print("Test accuracy:{:.2f}%".format(knn_acc*100))
```

```python
print(confusion_matrix(ytest, knn_predict))
print("\n")
print(classification_report(ytest, knn_predict))
```

Since the model has performed well on the test set, it doesnt require any hyperparameter tuning.

```python
svm = SVC()
svm.fit(xtrain, ytrain)
svm_predict = svm.predict(xtest)
svm_acc = accuracy_score(svm_predict, ytest)
print("Test accuracy:{:.2f}%".format(svm_acc*100))
```

```python
print(confusion_matrix(ytest, svm_predict))
print("\n")
print(classification_report(ytest, svm_predict))
```

```python
Rf = RandomForestClassifier()
Rf.fit(xtrain, ytrain)
Rf_predict = Rf.predict(xtest)

Rf_acc = accuracy_score(Rf_predict, ytest)
print("Test accuracy:{:.2f}%".format(Rf_acc*100))
```

```python
print(confusion_matrix(ytest, Rf_predict))
print("\n")
print(classification_report(ytest, Rf_predict))
```
